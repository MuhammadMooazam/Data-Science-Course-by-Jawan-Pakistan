# Question 01
#### Add two columns Speed_High_Low and HP_High_Low in the same dataset.
##### Dataset Link: https://www.kaggle.com/abcsds/pokemon

```python
import pandas as pd
```

```python
pkman = pd.read_csv("Pokemon.csv")
```

```python
pkman
```

```python
speedMean = pkman["Speed"].mean()
print(speedMean)

def setSpeedValue(speed):
    if speed<speedMean:
        return "Speed Low"
    elif speed==speedMean:
        return "Speed Neutral"
    elif speed>speedMean:
        return "Speed High"
    
pkman["Speed_High_Low"] = pkman["Speed"].apply(setSpeedValue)
```

```python
pkman
```

```python
hpMean = pkman["HP"].mean()
print(hpMean)

def setHpValue(hp):
    if hp<hpMean:
        return "HP Low"
    elif hp==hpMean:
        return "HP Neutral"
    elif hp>hpMean:
        return "HP High"
    
pkman["HP_High_Low"] = pkman["HP"].apply(setHpValue)
```
