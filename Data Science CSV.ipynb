# Feature Engineering

```python
pkman["Attack"].mean()
```

```python
attackMean = pkman["Attack"].mean()
print(attackMean)
```

```python
def setAttackValue(val):
    if val<attackMean:
        return "Attack Low"
    elif val==attackMean:
        return "Attakh Neutral"
    elif val>attackMean:
        return "Attack High"
```

```python
pkman["Attack_High_Low"] = pkman["Attack"].apply(setAttackValue)
```

```python
pkman
```

```python
spAttackMean = pkman["Sp. Atk"].mean()
print(spAttackMean)
```

```python
def setSpeedAttackValue(spatk):
    if spatk<spAttackMean:
        return "Speed Attack Low"
    elif spatk==spAttackMean:
        return "Speed Attack Neutral"
    elif spatk>spAttackMean:
        return "Speed Attack High"
```

```python
pkman["Speed Attack High Low"] = pkman["Sp. Atk"].apply(setSpeedAttackValue)
```
