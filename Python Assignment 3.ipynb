# Question 01
#### Create a new column in the provided dataset. Column name should be num_of_cast. You have to append or add a column in the dataset which contains the total number of cast in that particular movie or TV-Show.

```python
def cast_len(val):
    if val == "None":
        return 0
    else:
        return len(str(val).split(", "))
    
netflix["cast"].replace(np.nan, "None",inplace=True)
netflix["num_of_cast"] = netflix["cast"].apply(cast_len)
```
